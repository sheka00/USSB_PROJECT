# PPStructure

```python
import os
import cv2
from paddleocr import PPStructure
import matplotlib.pyplot as plt
from pytesseract import image_to_string
plt.rcParams['figure.figsize'] = (15, 20)
from google.colab.patches import cv2_imshow
```

```python
table_engine = PPStructure(table=False, ocr=False, show_log=True)
```

```python
img_path = '/content/Снимок.JPG'
img = cv2.imread(img_path)
result = table_engine(img)


for res in result:
  if res['type'] in ['text', 'reference']:
    x1, y1, x2, y2 = res['bbox']
    cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 0), 2)
    cv2.putText(img, res['type'], (x1, y1-2), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2_imshow(img[y1:y1+y2-y1, x1:x1+x2-x1])
    print(image_to_string(img[y1:y1+y2-y1, x1:x1+x2-x1], config='--psm=5', lang='rus'))
    print("--------------------------------------------------------------------------------------------------")

plt.imshow(img)
```
