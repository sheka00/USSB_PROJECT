```
plt.figure(figsize = (20,20))

aa = torch.load('/content/tensor.pt')
aa1 = torch.load('/content/tensor1.pt')

image = cv2.imdecode(np.fromfile('/content/unilm/dit/object_detection/publaynet_example.jpeg', dtype=np.uint8), cv2.IMREAD_COLOR) 
image_copy = image.copy()

for id,i in enumerate(aa):
  x1 = int(i[0].item())
  y1 = int(i[1].item())
  x2 = int(i[2].item())
  y2 = int(i[3].item())

  if aa1[id].item() == 0:
    cv2_imshow(image_copy[y1:y1+y2-y1, x1:x1+x2-x1])
    print(image_to_string(image_copy[y1:y1+y2-y1, x1:x1+x2-x1], config='--psm=5', lang='eng'))
    print("--------------------------------------------------------------------------------------------------")


  cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 1)
  cv2.putText(image, str(aa1[id].item()), (x1, y1-5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255))
  

plt.imshow(image)

#{0: "Text", 1: "Title", 2: "List", 3:"Table", 4:"Figure"}
```
