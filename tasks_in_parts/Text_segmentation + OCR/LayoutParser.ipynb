```python
import layoutparser as lp
import cv2
import matplotlib.pyplot as plt
from pytesseract import image_to_string
plt.rcParams['figure.figsize'] = (15, 20)
from google.colab.patches import cv2_imshow
```

```python
model = lp.Detectron2LayoutModel('lp://PubLayNet/faster_rcnn_R_50_FPN_3x/config', 
                                 extra_config=["MODEL.ROI_HEADS.SCORE_THRESH_TEST", 0.8],
                                 label_map={0: "Text", 1: "Title", 2: "List", 3:"Table", 4:"Figure"})
```

```python
image = cv2.imread('/content/Снимок.JPG')
layout = model.detect(image)


for i in layout:
  if i.type in ["Text", "List"]:
    x1 = int(i.block.x_1)
    y1 = int(i.block.y_1)
    x2 = int(i.block.x_2)
    y2 = int(i.block.y_2)

    cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 1)
    cv2.putText(image, i.type, (x1, y1-5), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 255, 0))
    cv2_imshow(image[y1:y1+y2-y1, x1:x1+x2-x1])
    print(image_to_string(image[y1:y1+y2-y1, x1:x1+x2-x1], config='--psm=5', lang='rus'))
    print("--------------------------------------------------------------------------------------------------")

plt.rcParams['figure.figsize'] = (10, 20)
plt.imshow(image)
```
